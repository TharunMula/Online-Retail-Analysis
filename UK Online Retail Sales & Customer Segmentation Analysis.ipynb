Transaction Type,
Calculating Revenue

```python
df['Transaction_Type']= df['Quantity'].apply(lambda x : 'Sales' if x >0 else 'Returns')
```

```python
df['Revenue']= df['Quantity']* df['UnitPrice']
```

Gross Revenue → Only positive sales

Return Amount → Absolute value of negative revenue

Net Revenue → Gross Revenue – Returns

Return Rate (%) → Returns / Gross Revenue

```python
gross_revenue = df[df['Quantity'] > 0]['Revenue'].sum()
```

```python
gross_revenue
```

```python
refund_amount= abs(df[df['Quantity'] < 0] ['Revenue'].sum())
```

```python
Net_Revenue= gross_revenue-refund_amount
```
